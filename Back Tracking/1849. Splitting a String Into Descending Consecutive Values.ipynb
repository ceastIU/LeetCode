1849. Splitting a String Into Descending Consecutive Values

Medium

You are given a string s that consists of only digits.

Check if we can split s into two or more non-empty substrings such that the numerical values of the substrings are in descending order and the difference between numerical values of every two adjacent substrings is equal to 1.

For example, the string s = "0090089" can be split into ["0090", "089"] with numerical values [90,89]. The values are in descending order and adjacent values differ by 1, so this way is valid.
Another example, the string s = "001" can be split into ["0", "01"], ["00", "1"], or ["0", "0", "1"]. However all the ways are invalid because they have numerical values [0,1], [0,1], and [0,0,1] respectively, all of which are not in descending order.
Return true if it is possible to split s​​​​​​ as described above, or false otherwise.

A substring is a contiguous sequence of characters in a string.

Example 1:

Input: s = "1234"
Output: false
Explanation: There is no valid way to split s.

Example 2:

Input: s = "050043"
Output: true
Explanation: s can be split into ["05", "004", "3"] with numerical values [5,4,3].
The values are in descending order with adjacent values differing by 1.

Example 3:

Input: s = "9080701"
Output: false
Explanation: There is no valid way to split s.

---
Observations:
- This follows a lot of the other problems that require chunking (creating a substring) and passing the remaining values. 
- Create every possible substring, and then do various forms of binding on that data

```python
class Solution:
    def splitString(self, s: str) -> bool:
        # this is the dfs code
        # passed in is the previous value, and the remaining s
        def dfs(prev, remaining):

            # if the length of the remaining string is zero we return false
            # this makes sure there is at least 2 seperate values as stated by the
            # problem: "two or more non-empty substrings"
            if len(remaining) == 0:
                return False

            # Next we directly check if the remaining s once converted to a int is 
            # 1 less then prev. If so we return true.
            # We know this to be valid because we only call recursive calls when
            # all previous values are descending and a difference of 1
            if int(remaining) == prev - 1:
                return True

            # If we reach this spot then it is time to iterate acorss the remaining
            # search space
            # We iterate from across the length of the remaining string
            for i in range(len(remaining)):

                # Create every single "chunk" of size i to check
                # Convert into an int
                chunk = int(remaining[:i+1])

                # Grab the remaining part of the string and MAKE SURE TO NAME IT 
                # DIFFERENTLY. This and chunk will be passed into the recursive call
                remaining2 = remaining[i+1:]

                # We want to make sure that the "chunk" we have satisfies the difference
                # condition of 1, otherwise there is no point in making futher recursive
                # calls. Also, since we initialize the recursive call below with Infinity
                # we make sure to still make calls when there is no "previous" value
                if chunk == prev - 1 or prev == float("inf"):

                    # Make the recursive call using the chunk as the previous value, and 
                    # passing in the new formed remaining2
                    # If the recursive call returns True, we make sure to return from this
                    # current recursive call as True
                    if dfs(chunk,remaining2):
                        return True

            # If we reach the end of the function without returning True, then there must
            # be no valid splitting, thus return False
            return False
        
        # Make the initial call to the DFS function using Infinity as the prev value, and 
        # the whole s as the "remaining"
        return dfs(float("inf"),s)
```

Fianlly a problem I was able to solve with relative ease. The time complexity is O(2^n) where n is the lenght of the string s. In the worst case, each character can be a part of a split or not, leading to 2 possibilities per character. The space complexity is O(n), where n is the length of the string and the space is the depth of the recursive calls.
