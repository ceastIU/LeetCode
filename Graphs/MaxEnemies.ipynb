Given an m x n matrix grid where each cell is either a wall 'W', an enemy 'E' or empty '0', return the maximum enemies you can kill using one bomb. You can only place the bomb in an empty cell.

The bomb kills all the enemies in the same row and column from the planted point until it hits the wall since it is too strong to be destroyed.

```python
from collections import deque
class Solution:
    def maxKilledEnemies(self, grid: List[List[str]]) -> int:
        rows, cols = len(grid), len(grid[0])
        ans = 0
        eRows = set()
        eCols = set()
        memoRow = {}
        memoCol = {}
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == "E":
                    eRows.add(i)
                    eCols.add(j)
        
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == "0" and (i in eRows or j in eCols):
                    # left
                    colE = 0
                    rowE = 0
                    found = False
                    if j in memoCol:
                        for key in memoCol[j]:
                            low, high = key
                            e = memoCol[j][key]
                            if low < i < high:
                                colE = e
                                found = True
                                break
                    if not found:
                        low = rows
                        high = 0
                        for dx, dy in [(-1,0),(1,0)]:
                            ti, tj = i, j
                            while 0 <= ti < rows and 0 <= tj < cols:
                                if grid[ti][tj] == "E":
                                    colE += 1
                                elif grid[ti][tj] == "W":
                                    break
                                ti += dx
                                tj += dy
                            low = min(ti, low)
                            high = max(ti,high)
                        
                        if j in memoCol:
                            memoCol[j][(low,high)] = colE
                        else:
                            memoCol[j] = {(low,high):colE}
                    found = False
                    if i in memoRow:
                        for key in memoRow[i]:
                            low, high = key
                            e = memoRow[i][key]
                            if low < j < high:
                                rowE = e
                                found = True
                                break
                    if not found:
                        low = cols
                        high = 0
                        for dx, dy in [(0,-1),(0,1)]:
                            ti, tj = i, j
                            while 0 <= ti < rows and 0 <= tj < cols:
                                if grid[ti][tj] == "E":
                                    rowE += 1
                                elif grid[ti][tj] == "W":
                                    break
                                ti += dx
                                tj += dy
                            low = min(tj, low)
                            high = max(tj,high)
                        
                        if i in memoRow:
                            memoRow[i][(low,high)] = rowE
                        else:
                            memoRow[i] = {(low,high):rowE}
                    ans = max(colE+rowE,ans)
        return ans
```
