There is an m x n rectangular island that borders both the Pacific Ocean and Atlantic Ocean. The Pacific Ocean touches the island's left and top edges, and the Atlantic Ocean touches the island's right and bottom edges.

The island is partitioned into a grid of square cells. You are given an m x n integer matrix heights where heights[r][c] represents the height above sea level of the cell at coordinate (r, c).

The island receives a lot of rain, and the rain water can flow to neighboring cells directly north, south, east, and west if the neighboring cell's height is less than or equal to the current cell's height. Water can flow from any cell adjacent to an ocean into the ocean.

Return a 2D list of grid coordinates result where result[i] = [ri, ci] denotes that rain water can flow from cell (ri, ci) to both the Pacific and Atlantic oceans.

```python
class Solution:
    def pacificAtlantic(self, heights: List[List[int]]) -> List[List[int]]:
        ROWS, COLS = len(heights), len(heights[0])
        pacfic = set()
        atlantic = set()
        res = []
        def dfs(r, c, path):
            # pacific is 0, 0
            if (r < 0 or c < 0) or (r,c) in pacfic:
                return "P"
            if (r == ROWS or c == COLS) or (r,c) in atlantic:
                return "A"
            # we have a cell that is within the grid looking to find whether or not it can reach both
            oceanP = False
            path.add((r,c))
            # can it reach the pacific?
            pacficDir = [(-1,0),(0,1),(0,-1)]
            for xx, yy in pacficDir:
                newR, newC = r + xx, c + yy
                if heights[newR][newC] <= heights[r][c] and (newR, newC) not in path:
                    if dfs(newR,newC,path) == "P":
                        pacfic.add((r,c))
                        oceanP = True
            oceanA = False
            # can it reach the pacific?
            atlanticDir = [(1,0),(0,1),(0,-1)]
            for xx, yy in atlanticDir:
                newR, newC = r + xx, c + yy
                if heights[newR][newC] <= heights[r][c] and (newR, newC) not in path:
                    if dfs(newR,newC,path) == "A":
                        atlantic.add((r,c))
                        oceanA = True
            if oceanA and oceanP:
                res.append([r,c])
        for i in range(ROWS):
            for j in range(COLS):
                dfs(i,j,set())
        return res
            
```

```python
# Round 3
# the key idea here is to start the edges of the ocean and move inward
# We do this in pairs every row of the west and east side of the island
# and Every col of the north and south side of the island
# Each DFS passing in the correct SET() that represents an ocean
class Solution:
    def pacificAtlantic(self, heights: List[List[int]]) -> List[List[int]]:
        ROWS, COLS = len(heights), len(heights[0])
        pacific = set()
        atlantic = set()
        res = []
        # define the dfs that moves from edge to center
        def dfs(r,c,ocean,prevHeight):
            # Check to see if:
            # - we are back in the ocean
            # - we are in visit
            # - we have moved downhill
            if not (0 <= r < ROWS and 0 <= c < COLS) or (r,c) in ocean or heights[r][c] < prevHeight:
                return
            # If we reach this point then we have a valid cell that can reach the passed in ocean
            ocean.add((r,c))
            for xx, yy in [(-1,0),(1,0),(0,-1),(0,1)]:
                newR, newC = r + xx, c + yy
                dfs(newR, newC, ocean, heights[r][c])
            
        for i in range(ROWS):
            dfs(i,0,pacific,0)
            dfs(i, COLS-1,atlantic,0)
        
        for i in range(COLS):
            dfs(0,i, pacific,0)
            dfs(ROWS-1,i,atlantic,0)

        for i in range(ROWS):
            for j in range(COLS):
                if (i,j) in pacific and (i,j) in atlantic:
                    res.append([i,j])
        return res
```

Time Complexity: The main time complexity comes from the two DFS traversals that you perform, one for the cells that can reach the Pacific Ocean and the other for the cells that can reach the Atlantic Ocean. Each DFS can potentially visit every cell in the grid once.

The first two nested loops that initiate the DFS from the edge cells (i, 0 and i, COLS-1 for rows and 0, i and ROWS-1, i for columns) take O(ROWS + COLS) time, where ROWS is the number of rows and COLS is the number of columns in the grid.

The final nested loop that iterates through all cells in the grid (for i in range(ROWS) and for j in range(COLS)) takes O(ROWS * COLS) time.

Since each DFS traversal potentially visits every cell once, the total time complexity of the code is O(ROWS * COLS).

Space Complexity: The space complexity is determined by the space needed for the sets pacific and atlantic to keep track of cells that can reach the respective oceans. In the worst case, all cells in the grid can be added to these sets.

The pacific and atlantic sets can each store up to O(ROWS * COLS) cells, resulting in a total space complexity of O(ROWS * COLS).
Overall, your code has a time complexity of O(ROWS * COLS) and a space complexity of O(ROWS * COLS).
