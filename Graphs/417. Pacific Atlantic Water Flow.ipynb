417. Pacific Atlantic Water Flow

Medium

There is an m x n rectangular island that borders both the Pacific Ocean and Atlantic Ocean. The Pacific Ocean touches the island's left and top edges, and the Atlantic Ocean touches the island's right and bottom edges.

The island is partitioned into a grid of square cells. You are given an m x n integer matrix heights where heights[r][c] represents the height above sea level of the cell at coordinate (r, c).

The island receives a lot of rain, and the rain water can flow to neighboring cells directly north, south, east, and west if the neighboring cell's height is less than or equal to the current cell's height. Water can flow from any cell adjacent to an ocean into the ocean.

Return a 2D list of grid coordinates result where result[i] = [ri, ci] denotes that rain water can flow from cell (ri, ci) to both the Pacific and Atlantic oceans.

---
Observations:
- This is a far more challenging problem and I love it
- There could be two set for cells that can flow to the Pacific or the atlantic. Use those data structures to make later recursive calls faster.
- The hard part is how to set up this recursive function to figure out whether a cell flows to both pacific and atlantic


```python
class Solution:
    def pacificAtlantic(self, heights: List[List[int]]) -> List[List[int]]:
        ROWS, COLS = len(heights), len(heights[0])
        pacfic = set()
        atlantic = set()
        res = []
        def dfs(r, c, path):
            # pacific is 0, 0
            if (r < 0 or c < 0) or (r,c) in pacfic:
                return "P"
            if (r == ROWS or c == COLS) or (r,c) in atlantic:
                return "A"
            # we have a cell that is within the grid looking to find whether or not it can reach both
            oceanP = False
            path.add((r,c))
            # can it reach the pacific?
            pacficDir = [(-1,0),(0,1),(0,-1)]
            for xx, yy in pacficDir:
                newR, newC = r + xx, c + yy
                if heights[newR][newC] <= heights[r][c] and (newR, newC) not in path:
                    if dfs(newR,newC,path) == "P":
                        pacfic.add((r,c))
                        oceanP = True
            oceanA = False
            # can it reach the pacific?
            atlanticDir = [(1,0),(0,1),(0,-1)]
            for xx, yy in atlanticDir:
                newR, newC = r + xx, c + yy
                if heights[newR][newC] <= heights[r][c] and (newR, newC) not in path:
                    if dfs(newR,newC,path) == "A":
                        atlantic.add((r,c))
                        oceanA = True
            if oceanA and oceanP:
                res.append([r,c])
        for i in range(ROWS):
            for j in range(COLS):
                dfs(i,j,set())
        return res
            
```

```python
# Round 2: With the new understanding of working backwards
class Solution:
    def pacificAtlantic(self, heights: List[List[int]]) -> List[List[int]]:
        ROWS, COLS = len(heights), len(heights[0])
        pacific = set()
        atlantic = set()
        res = []
        # dfs: ( row, col, set we are adding to, previous height )
        def dfs(r,c,visit,prevHeight):
            # we want to make sure we aren't back in the ocean, or we aren't in the visited cell, 
            # or we aren't travel down when we want to travel up
            if (r,c) in visit or r < 0 or c < 0 or r == ROWS or c == COLS or heights[r][c] < prevHeight:
                return

            # otherwise we are a valid node
            visit.add((r,c))
            for xx, yy in [(-1,0),(1,0),(0,-1),(0,1)]:
                newR, newC = r + xx, c + yy
                dfs(newR,newC,visit,heights[r][c])

        for i in range(COLS):
            dfs(0,i,pacific,0)
            dfs(ROWS-1,i,atlantic,0)

        for i in range(ROWS):
            dfs(i,0,pacific,0)
            dfs(i,COLS-1,atlantic,0)
        print(pacific,atlantic)     

```
