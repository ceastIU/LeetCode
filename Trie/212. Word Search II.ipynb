212. Word Search II

Hard

Given an m x n board of characters and a list of strings words, return all words on the board.

Each word must be constructed from letters of sequentially adjacent cells, where adjacent cells are horizontally or vertically neighboring. The same letter cell may not be used more than once in a word.

---

Example 1:


Input: board = [["o","a","a","n"],["e","t","a","e"],["i","h","k","r"],["i","f","l","v"]], words = ["oath","pea","eat","rain"]
Output: ["eat","oath"]

```python
class TrieNode:
    def __init__(self) -> None:
        self.children = {}

class Solution:
    def __init__(self) -> None:
        self.root = TrieNode()
        
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        starting = {}
        for word in words:
            starting[word[0]] = len(word)
        # construct the trie tree through dfs
        def dfs(root, cor,depth, path):
            if len(path) == depth + 1:
                print(path)
                return
            cur = root
            i, j = cor
            c = board[i][j]
            print(c)
            if c not in cur.children:
                cur.children[c] = TrieNode()
            cur = cur.children[c]
            directions = [(-1,0),(1,0),(0,-1),(0,1)]
            path.add((i,j))
            for xx, yy in directions:
                iIdx = i + xx
                jIdx = j + yy

                if 0 <= iIdx < len(board) and 0 <= jIdx < len(board[0]) and (iIdx, jIdx) not in path:
                    
                    dfs(cur,(iIdx,jIdx),depth,path)
            path.remove((i,j))

                
        for i in range(len(board)):
            for j in range(len(board[0])):
                if board[i][j] in starting:
                    path = set()
                    path.add((i,j))
                    dfs(self.root, (i,j), starting[board[i][j]], path)
        solutions = []
        for word in words:
            cur = self.root
            flag = True
            for c in word:
                if c not in cur.children:
                    flag = False
                    break
                cur = cur.children[c]
            if flag:
                solutions.append(word)
        return solutions
```

```python
class TrieNode:
    def __init__(self) -> None:
        self.children = {}
        self.refs = 0 
        self.end = False

    def addWord(self,word):
        node = self
        node.refs += 1
        for c in word:
            if c not in node.children:
                node.children[c] = TrieNode
            node = node.children[c]
            node.refs += 1
        node.end = True

    def removeWord(self,word):
        node = self
        node.refs -= 1
        for c in word:
            if c in node.children:
                node = node.children[c]
                node.refs -= 1
        node.end = False

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        rows, cols = len(board), len(board[0])
        tree = TrieNode()
        for word in words:
            tree.addWord(word)
        res = set()
        visit = set()
        def dfs(i,j,node,word):
            if not (0 <= i < rows and 0 <= j < cols) or board[i][j] not in node.children or node.children[board[i][j]].refs < 1:
                return
            word += board[i][j]
            visit.add((i,j))
            node = node.children[board[i][j]]
            if node.end:
                res.add(word)
                tree.removeWord(word)
                node.end = False
            
            for dx, dy in [(-1,0),(1,0),(0,-1),(0,1)]:
                xx, yy = i + dx, j + dy
                if (xx, yy) not in visit:
                    dfs(xx,yy,node,word)
        for i in range(rows):
            for j in range(cols):
                dfs(i,j,tree,"")
        return list(res)
                
```
